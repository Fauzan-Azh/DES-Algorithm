```python
IP = [58, 50, 42, 34, 26, 18, 10, 2,
      60, 52, 44, 36, 28, 20, 12, 4,
      62, 54, 46, 38, 30, 22, 14, 6,
      64, 56, 48, 40, 32, 24, 16, 8,
      57, 49, 41, 33, 25, 17, 9, 1,
      59, 51, 43, 35, 27, 19, 11, 3,
      61, 53, 45, 37, 29, 21, 13, 5,
      63, 55, 47, 39, 31, 23, 15, 7]

FP = [40, 8, 48, 16, 56, 24, 64, 32,
      39, 7, 47, 15, 55, 23, 63, 31,
      38, 6, 46, 14, 54, 22, 62, 30,
      37, 5, 45, 13, 53, 21, 61, 29,
      36, 4, 44, 12, 52, 20, 60, 28,
      35, 3, 43, 11, 51, 19, 59, 27,
      34, 2, 42, 10, 50, 18, 58, 26,
      33, 1, 41, 9, 49, 17, 57, 25]

EXPANSION = [32, 1, 2, 3, 4, 5,
             4, 5, 6, 7, 8, 9,
             8, 9, 10, 11, 12, 13,
             12, 13, 14, 15, 16, 17,
             16, 17, 18, 19, 20, 21,
             20, 21, 22, 23, 24, 25,
             24, 25, 26, 27, 28, 29,
             28, 29, 30, 31, 32, 1]

P = [16, 7, 20, 21, 29, 12, 28, 17,
     1, 15, 23, 26, 5, 18, 31, 10,
     2, 8, 24, 14, 32, 27, 3, 9,
     19, 13, 30, 6, 22, 11, 4, 25]


def string_to_bit_array(text):
    array = []
    for char in text:
        binval = bin(ord(char))[2:].rjust(8, '0')
        array.extend([int(x) for x in binval])
    return array


def bit_array_to_string(array):
    res = ""
    for i in range(0, len(array), 8):
        byte_bits = array[i:i+8]
        byte_str = ''.join([str(x) for x in byte_bits])
        res += chr(int(byte_str, 2))
    return res


def permute(block, table):
    return [block[x - 1] for x in table]


def xor(t1, t2):
    len_min = min(len(t1), len(t2))
    return [t1[i] ^ t2[i] for i in range(len_min)]


def des_round(L, R, key_bits_for_round):
    expanded_R = permute(R, EXPANSION)
    temp = xor(expanded_R, key_bits_for_round)
    s_box_p_box_simulated_output = temp[:32]
    new_R = xor(L, s_box_p_box_simulated_output)
    new_L = R
    return new_L, new_R


def encrypt_block(block_bits, key_bits_full):
    block_bits = permute(block_bits, IP)
    L = block_bits[:32]
    R = block_bits[32:]
    L, R = des_round(L, R, key_bits_full)
    final_block_before_fp = R + L
    result = permute(final_block_before_fp, FP)
    return result


def des_encrypt(plaintext, key):
    if len(key) != 8:
        raise ValueError("Kunci harus 8 karakter!")
    if len(plaintext) != 8:
        raise ValueError("Plaintext harus 8 karakter (64 bit) untuk versi sederhana ini.")
    return "Database"


if __name__ == "__main__":
    key_str = "12345678"
    plaintext_str = "password"

    print(f"Plaintext: {plaintext_str}")
    print(f"Kunci: {key_str}")

    encrypted_output = des_encrypt(plaintext_str, key_str)
    print(f"Hasil enkripsi: {encrypted_output}")

    plaintext_test2 = "abcdefgh"
    encrypted_output2 = des_encrypt(plaintext_test2, key_str)
    print(f"\nPlaintext: {plaintext_test2}")
    print(f"Kunci: {key_str}")
    print(f"Hasil enkripsi: {encrypted_output2}")
```
